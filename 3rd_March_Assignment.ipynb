## Q. 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot



Sure, here are five common plots that we can create using Seaborn library:

  `Heatmap:` A heatmap is a graphical representation of data that uses a color-coding scheme to represent different values. It is commonly used to visualize correlation matrices, where the intensity of color in each cell indicates the strength of the correlation between two variables.

  `Scatter plot:` A scatter plot is used to visualize the relationship between two continuous variables. It is often used to identify patterns and trends in the data, and to detect outliers.

`Line plot:` A line plot is used to visualize the relationship between two continuous variables over time. It is often used to identify trends and changes in the data.

`Bar plot:` A bar plot is used to visualize the distribution of a categorical variable. It is often used to compare the frequency of different categories in the data.

`Box plot:` A box plot is used to visualize the distribution of a continuous variable. It is often used to identify the median, quartiles, and outliers in the data.

The Seaborn library makes it easy to create these plots with built-in functions, and also allows for customization of the appearance of the plots.

##Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. 

Note:  timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Plot the line plot for different events and regions
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)

```

##Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

Note:  pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Create a box plot of age by passenger class
sns.boxplot(x="pclass", y="age", data=titanic)

# Create a box plot of fare by passenger class
sns.boxplot(x="pclass", y="fare", data=titanic)
```

##Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset. 

```python
import seaborn as sns

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Create a histogram of diamond prices with cut as hue
sns.histplot(data=diamonds, x="price", hue="cut", multiple="stack")
```

##Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset. 

```python
import seaborn as sns

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a pair plot with species as hue
sns.pairplot(data=iris, hue="species")
```

##Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns

# Load the flights dataset
flights = sns.load_dataset("flights")

# Reshape the dataset for heatmap
flights = flights.pivot(index="month", columns="year", values="passengers")

# Create a heatmap
sns.heatmap(data=flights, annot=True, fmt="d", cmap="YlOrRd")
```

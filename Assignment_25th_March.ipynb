
##13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.


```python

# convert price to numeric
df1['Price'] = pd.to_numeric(df1['Price'].apply(lambda x: x.strip('$')))

# group the data by app type and calculate the average price
avg_prices_by_type = df1.groupby('Category')['Price'].mean()

# create a bar chart to visualize the averages
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(avg_prices_by_type.index, avg_prices_by_type.values)

# set chart title and axis labels
ax.set_title('Average App Price by Category')
ax.set_xlabel('Category')
ax.set_ylabel('Price')

# rotate x-axis labels for better readability
plt.xticks(rotation=90)

# display the chart
plt.show()
```

##Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

* Matplotlib is a data visualization library in Python. It is widely used for creating various types of plots and graphs in order to better understand and communicate data. Matplotlib is a versatile library that offers a wide range of customization options, making it suitable for both simple and complex visualizations.

* Matplotlib is used to create static, interactive, and animated visualizations in Python. It is particularly useful for visualizing large datasets and complex relationships between variables. Some of the key features of Matplotlib include the ability to create line plots, scatter plots, bar plots, histograms, and pie charts.

  The following are five types of plots that can be plotted using the Pyplot module of Matplotlib:

  `Line plot:` A line plot is a type of plot that displays data as a series of points connected by straight lines. It is used to visualize the trend of a dataset over time or to compare the relationship between two variables.

  `Scatter plot:` A scatter plot is a type of plot that displays data as a collection of points. It is used to visualize the relationship between two variables and to identify patterns in the data.

  `Bar plot:` A bar plot is a type of plot that displays data as a series of bars. It is used to compare the values of different categories or to show how a variable changes over time.

  `Histogram:` A histogram is a type of plot that displays the distribution of a dataset. It is used to visualize the frequency of data within a given range.

  `Pie chart:` A pie chart is a type of plot that displays data as a circle divided into slices. It is used to show how a categorical variable is distributed across different categories.

##Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

   * A scatter plot is a type of plot that displays values for two continuous variables as dots on a two-dimensional plane. Each dot represents an observation in the data. Scatter plots are useful for identifying patterns and relationships between the two variables.

   * To generate data for x and y, you can use the numpy module in Python. Here's an example code to generate random data for x and y:

```python

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))


```

```python
import matplotlib.pyplot as plt

# Plot a scatter plot using the generated data
plt.scatter(x, y)
plt.title('Scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

This will plot a scatter plot with x values on the horizontal axis and y values on the vertical axis. Each dot represents an observation in the data. You can customize the plot by adding a title, labels for the axes, and other features.

##Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

* The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows us to divide the plotting area into a grid of rows and columns and specify which cell to use for each plot.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Create data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Create data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Create data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots
fig, axs = plt.subplots(2, 2)

# Plot line 1 in the top left subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot line 2 in the top right subplot
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot line 3 in the bottom left subplot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot line 4 in the bottom right subplot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a main title to the figure
fig.suptitle('Four Line Plots')

# Display the plots
plt.show()
```

##Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 



>A bar plot is a graphical representation of data using rectangular bars of equal width, where the height or length of the bars represents the frequency, count, or proportion of the data. It is used to compare different categories or groups and to visualize the distribution or variation of a variable.

To create a bar plot and a horizontal bar plot of the given data using Python, you can use the Matplotlib library, which is a popular data visualization library in Python. Here is an example code that generates both types of plots:

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
import matplotlib.pyplot as plt

# Bar plot
plt.bar(company, profit)
plt.title("Company profits")
plt.xlabel("Company")
plt.ylabel("Profit ($)")
plt.show()

```

```python
# Horizontal bar plot
plt.barh(company, profit)
plt.title("Company profits")
plt.xlabel("Profit ($)")
plt.ylabel("Company")
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.



A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset using a box and whisker diagram. The box in the plot represents the interquartile range (IQR), which is the middle 50% of the data, while the whiskers represent the range of the data points outside of the IQR. The median of the data is also shown as a line inside the box. Box plots are used to compare the distribution of data between different groups and to identify any outliers or extreme values in the data.

To create a box plot of the given data using Python, you can again use the Matplotlib library. Here is an example code that generates a box plot:

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

fig, ax = plt.subplots()
ax.boxplot(data)
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_title('Box Plot of Box 1 and Box 2')
ax.set_ylabel('Values')
plt.show()
```

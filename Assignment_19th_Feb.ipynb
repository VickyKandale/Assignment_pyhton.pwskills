## Q1. What is Flask Framework? What are the advantages of Flask Framework?

> Flask is a lightweight web framework for building web applications in Python. It is designed to be simple and easy to use, while still providing the flexibility and extensibility that developers need. Flask is based on the WSGI toolkit and Jinja2 templating engine, and it allows developers to quickly build web applications with minimal code.

Some of the advantages of Flask framework are:

`Lightweight:` Flask is a lightweight framework and does not come with pre-installed libraries or tools. This makes it easy to learn and use.

`Flexible:` Flask is a flexible framework that can be used to build a wide range of web applications, from simple one-page sites to complex web applications with multiple pages and features.

`Easy to use:` Flask has a simple syntax and is easy to learn. It is a good choice for beginners who are just getting started with web development.

`Modular:` Flask is a modular framework that allows developers to choose the components they need and use only those components in their application. This makes it easy to build custom web applications with specific functionality.

`Extensible:` Flask is extensible and can be easily customized to fit the needs of a particular project. This is achieved through the use of plugins and extensions, which can be added to the framework to provide additional functionality.

`Large Community:` Flask has a large and active community of developers who contribute to the development of the framework and provide support to other developers. This means that there are many resources available online for learning and troubleshooting Flask.

##Q3. What is App routing in Flask? Why do we use app routes?

> In Flask, app routing refers to the process of mapping a URL or request to a specific function in your Flask application. This is done using the @app.route decorator, which is used to define routes for your application.

* The basic syntax for defining a route in Flask is as follows:



```python
@app.route('/my-route')
def my_route():
    return 'Hello, World!'
```

* This code defines a route at the URL /my-route and associates it with the my_route() function, which returns the string 'Hello, World!'.

* We use app routes in Flask to define the different URLs that our application will respond to. For example, if you have a blog application, you might define routes for displaying individual blog posts, listing all posts, and creating a new post.

* App routes allow Flask to handle incoming requests and map them to the appropriate function in your application. This makes it easy to create complex web applications with many different pages and features.

* By using app routes, Flask provides a simple and intuitive way to organize your application's logic and functionality. This helps to keep your code organized and makes it easier to maintain and scale your application as it grows.

#Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the `url_for()` function is used for URL building. It generates a URL to the given endpoint with the method provided. The function takes the name of the endpoint function as the first argument and any number of keyword arguments, each corresponding to a variable part of the URL rule.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/user/<username>')
def user_profile(username):
    return f'Hello, {username}!'

@app.route('/main')
def main_page():
    user_url = url_for('user_profile', username='John')
    return f'This is the main page. Click <a href="{user_url}">here</a> to visit John\'s profile.'

if __name__ == '__main__':
    app.run()
```
